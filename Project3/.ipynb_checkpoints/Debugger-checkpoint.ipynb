```python
weightMap = eval(firstTest.loc['OutputNode', 'weightMap'])
output = 0

print(weightMap)

for key, index in weightMap.items():
    if key == 'Intercept':
        newValue = 1
        output += index * 1
    else:
        newValue = index * eval(firstTest.loc[key, 'currentOutputs'])[1]
        output += newValue
        
print(output)

otherOutput = 1 / (1 + math.exp(-output))

print(otherOutput)
```
