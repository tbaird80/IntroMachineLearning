```python
def bresenham_line(x0, y0, x1, y1):
    # Initialize the direction of the line
    dx = abs(x1 - x0)
    sx = 1 if x0 < x1 else -1
    dy = abs(y1 - y0)
    sy = 1 if y0 < y1 else -1
    error = dx + dy

    if dx > dy:
        # Initialize decision variables for |slope| <= 1
        err = dx / 2.0
        while x0 != x1:
            plot(x0, y0)
            err -= dy
            if err < 0:
                y0 += sy
                err += dx
            x0 += sx
    else:
        # Initialize decision variables for |slope| > 1
        err = dy / 2.0
        while y0 != y1:
            plot(x0, y0)
            err -= dx
            if err < 0:
                x0 += sx
                err += dy
            y0 += sy

    # Plot the final point
    plot(x0, y0)

def plot(x, y):
    print(f"Plotting point at ({x}, {y})")
```

```python
def wiki_plot(x0, y0, x1, y1):
    # Initialize the direction of the line
    dx = abs(x1 - x0)
    sx = 1 if x0 < x1 else -1
    dy = -1 * abs(y1 - y0)
    sy = 1 if y0 < y1 else -1
    error = dx + dy

    while True:
        plot(x0, y0)
        if x0 == x1 and y0 == y1:
            break
        e2 = 2 * error
        if e2 >= dy:
            if x0 == x1:
                break
            error = error + dy
            x0 = x0 + sx
        if e2 <= dx:
            if y0 == y1:
                break
            error = error + dx
            y0 = y0 + sy

def plot(x, y):
    print(f"Plotting point at ({x}, {y})")
```

```python
def wiki_plot2(x0, y0, x1, y1):
    if abs(y1 - y0) < abs(x1 - x0):
        if x0 > x1:
            plotLineLow(x1, y1, x0, y0)
        else:
            plotLineLow(x0, y0, x1, y1)
    else:
        if y0 > y1:
            plotLineHigh(x1, y1, x0, y0)
        else:
            plotLineHigh(x0, y0, x1, y1)
            
def plotLineLow(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    yi = 1
    if dy < 0:
        yi = -1
        dy = -dy
    D = (2 * dy) - dx
    y = y0
    xStep = 1 if x0 < x1 else -1
    
    for x in range(x0, x1+1, xStep):
        plot(x, y)
        if D > 0:
            y = y + yi
            D = D + (2 * (dy - dx))
        else:
            D = D + 2*dy
            
def plotLineHigh(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    xi = 1
    if dx < 0:
        xi = -1
        dx = -dx
    D = (2 * dx) - dy
    x = x0
    yStep = 1 if y0 < y1 else -1
    
    for y in range(y0, y1+1, yStep):
        plot(x, y)
        if D > 0:
            x = x + xi
            D = D + 2*dx
        else:
            D = D + 2*dx
            
```
